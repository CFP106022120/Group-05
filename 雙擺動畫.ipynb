```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation

%matplotlib notebook

def update_points_spring(num):
    
    if point_ani_spring.set_data(x1[num], y1[num]) :
        return point_ani_spring,
        
    elif point_ani2_spring.set_data(x2[num], y2[num]) :
        return point_ani2_spring,
    
    elif line1_spring.set_data([x11[num],x12[num]],[y11[num],y12[num]]) :
        return line1_spring,
    
    elif line2_spring.set_data([x21[num],x22[num]],[y21[num],y22[num]]) :
        return line2_spring,


M1=3
M2=7
L1=3
L2=3
g=9.8
theta10=0*np.pi/180
theta20=1*np.pi/180
theta10dot=0.2
theta20dot=0.2
wa=(g*M2/M1/L2)**(1/2)
wb=(g/M1/L2)**(1/2)
j=(-1)**0.5
B1=(j*wb*theta20+theta20dot)/(j*wb-j*wa)/M2/L2
B2=(j*wa*theta10+theta10dot)/(j*wa-j*wb)/L2
b1=(-j*wb*theta20+theta20dot)/(-j*wb+j*wa)/M2/L2
b2=(-j*wa*theta10+theta10dot)/(-j*wa+j*wb)/L2

t= np.linspace(0, 10*np.pi, 500)
    
theta1=B1*(M1+M2)*L1*np.exp((-1)*j*wa*t)+B2*L2*np.exp((-1)*j*wb*t)+b1*(M1+M2)*L1*np.exp(j*wa*t)+b2*L2*np.exp(j*wb*t)
theta2=B1*M2*L2*np.exp((-1)*j*wa*t)+B2*(L1+M1*L2)*np.exp((-1)*j*wb*t)+b1*M2*L2*np.exp(j*wa*t)+b2*(L1+M1*L2)*np.exp(j*wb*t)
    
x1=L1*np.sin(theta1)
x11=[0]*100
x12=L1*np.sin(theta1)
x2=x1+L2*np.sin(theta2)
x21=L1*np.sin(theta1)
x22=x1+L2*np.sin(theta2)
    
y1=-L1*np.cos(theta1)
y11=[0]*100
y12=-L1*np.cos(theta1) 
y2=y1-L2*np.cos(theta2)
y21=-L1*np.cos(theta1)
y22=y1-L2*np.cos(theta2)
    
    
fig = plt.figure(tight_layout=True)
#plt.plot(x11,y1,x22,y2)
point_ani_spring, = plt.plot(x11[0],y11[0], "ro")
point_ani2_spring, = plt.plot(x22[0],y22[0], "go")

line1_spring, = plt.plot([x11[0],x21[0]],[y11[0],y21[0]],color="black",ls=":")
line2_spring, = plt.plot([x12[0],x22[0]],[y12[0],y22[0]],color="black",ls=":")


plt.xlabel("x")
plt.ylabel("y")

plt.xlim(-10,10)
plt.ylim(-50,50)
#print(min(ylabel.real.tolist()))
#print(max(ylabel.real.tolist()))

plt.grid(ls="--")

ani_spring = animation.FuncAnimation(fig, update_points_spring, np.arange(0, 100), interval=100, blit=True)

plt.show()
```
